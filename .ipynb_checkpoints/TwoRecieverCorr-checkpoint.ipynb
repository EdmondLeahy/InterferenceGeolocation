```python
import pandas as pd
import numpy as np
from scipy.fft import fft
import matplotlib.pyplot as plt
import random 
from scipy import signal
```

```python
def calc_corr(data1, data2, tau):
    if tau >= 0:
            xx = data1
            yy = data2
            tau = tau
    else:
        xx = data2
        yy = data1
        tau = -tau
            
    yy = np.append(yy[tau:len(xx)], np.zeros(tau))
    return np.matmul(np.transpose(xx),yy) /(len(data1)-tau)
    
    
```

```python
def cross_corr(data1, data2, corr_range=1.0):
    corr = []
    size = int(len(data1)*corr_range)
    for lag in range(-size,size):
        corr.append(calc_corr(data1, data2, lag))
    
    return np.array(corr)
```

```python
def run_np_corr(series_rx1, series_rx2):
    np_corr = abs(signal.correlate(series_rx1, series_rx2))/len(series_rx1)
    data_size = int(len(np_corr)/2)
    np_corr = np.array(np_corr)
    max_val = np.max(np_corr)
    max_loc = np.argmax(np_corr)
    average = np.average(np_corr)
    std = np.std(np_corr)

    if max_val > (average + 10*std):
        return_val = (np_corr, True, max_val, average, max_val/average)
    else:
        return_val = (np_corr, False, max_val, average, max_val/average)
    
    return return_val
```

```python
def get_epoch_series(spkrlr_data, epoch):
    series =[]
    for log in spkrlr_data:
        if float(log[6]) == epoch:
            series.extend(get_series_from_log(log))
   
    return series
```

```python
def get_series_from_log(log):
    series = [int(dat) for dat in log[12:-1]]
    return series
```

```python
def get_offset_between_rx(rx1,rx2):
    base_time = float(rx1[60][6]) # Check 1 min in to file in case not perfect overlap
    for log in rx2:
        check_time = float(log[6])
        if base_time == check_time:
            try:
                offset = int(check_time - base_time)
            except:
                raise UserWarning('non integer offset. Sorry')
            return offset
        
    return -1
```

```python
def split_i_q(data):
    i=[]
    q=[]
    for j in range(len(data)):
        if j%2 == 0:
            i.append(data[j])
        else:
            q.append(data[j])
    return (np.array(i), np.array(q))
```
