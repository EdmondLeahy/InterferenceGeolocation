```python
import numpy as np
from scipy import signal
import numpy.random as rnd
import matplotlib.pyplot as plt
```

```python
array_size = 10000
shift = 1000
window_size = 2000
sig = np.random.randint(0,100,array_size)

# zero pad the second signal
window_loc = int((array_size - window_size)/2)
leading_zeros = np.zeros(window_loc + shift)
trailing_zeros = np.zeros(window_loc - shift)
sig2 = np.append(leading_zeros,sig[window_loc:window_loc+window_size])
sig2 = np.append(sig2, trailing_zeros)
```

```python
plt.plot(sig)
plt.plot(sig2)
```

```python
corr = signal.correlate (sig, sig2)

corr_lags = signal.correlation_lags(len(sig), len(sig2))

corr /= np.max(corr)
```

```python
plt.plot(corr_lags,corr)
```

```python
corr_lags[np.argmax(corr)]
```

# Do the same, but with Sin wave

```python
array_size = 1000
shift = 0
window_size = 500
t = np.arange(1,array_size, 1)
sig = np.cos(1/10*t)

# zero pad the second signal
window_loc = int((array_size - window_size)/2)
leading_zeros = np.zeros(window_loc + shift)
trailing_zeros = np.zeros(window_loc - shift)
sig2 = np.append(leading_zeros,sig[window_loc:window_loc+window_size])
sig2 = np.append(sig2, trailing_zeros)
```

```python
var = 0.01
sig = sig + np.sqrt(var) * rnd.randn(len(sig))
sig2 = sig2 + np.sqrt(var) * rnd.randn(len(sig2))
```

```python
plt.plot(sig)
plt.plot(sig2)
```

```python
corr = signal.correlate (sig, sig2)

corr_lags = signal.correlation_lags(len(sig), len(sig2))

corr /= np.max(corr)
plt.plot(corr_lags,corr)
```
